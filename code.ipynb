```python
import urllib.request
url = "https://raw.githubusercontent.com/karpathy/char-rnn/master/data/tinyshakespeare/input.txt"
urllib.request.urlretrieve(url , "input.txt")
print("Dataset loaded!")
```

```python
with open('input.txt' , 'r' , encoding='utf-8') as f:
    text = f.read()
```

```python
chars = sorted(list(set(text)))
print(''.join(chars))
size=len(chars)
print(size)
```

```python
stoi = {ch:i for i , ch in enumerate(chars)} # Create a dictionary a:0
itos= {i:ch for i , ch in enumerate(chars)}  #create a dictionary 0:a

def encode(s): #  this func  take a string and output a list of integers
    return [stoi[c] for c in s]
def decode(s):
    return (''.join([itos[c] for c in s]))
```
